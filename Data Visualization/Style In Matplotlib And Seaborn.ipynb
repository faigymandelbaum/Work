```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.style as style
import seaborn as sns
```

```python
donuts = pd.read_csv('donuts.csv')
```

```python
donuts.head()
```

```python
style.available
```

```python
donuts['branch'].value_counts().plot.bar()
donuts['type'].value_counts().plot.bar()
plt.suptitle('Donut Sales', size = 'xx-large', weight = 'bold')
plt.xlabel('Flavor')
plt.ylabel('Sales')
style.use('seaborn-paper')
```

This chart is suitable for a baked goods wholesaler that sells baked goods to groceries around the city.

```python
donuts['branch'].value_counts().plot.bar()
donuts['type'].value_counts().plot.bar()
plt.title('Donut Sales', size = 'xx-large')
plt.xlabel('Flavor')
plt.ylabel('Sales')
style.use('bmh')
```

The chart above puts out clearly the amounts sold for every flavor. It is imperative for the company to know exactly how many donuts are sold in each flavor.

```python
sns.catplot(x= 'branch', hue = 'type', kind ='count', data = donuts)
sns.set_style('white')
sns.set_palette('colorblind')
sns.set_context('poster')
```

The histogram above is good as an advertisement for the visually challenged and older people that are looking to treat families that they visit. These colors are most easily discerned by the eye. In addition, the graph is large and written in bold.

```python
sns.catplot(x= 'branch', hue = 'type', kind ='count', data = donuts)
style.use('seaborn-notebook')
sns.set_palette('bright')
sns.set_context('notebook')
```

This chart is good for a teacher that wants to show her students how charts give clarity.

```python
sns.catplot(x= 'branch', hue = 'type', kind ='count', data = donuts)
style.use('seaborn-notebook')
sns.set_palette('bright')
sns.set_context('notebook')
plt.text(x = 0.7, y = 42.05, s ='Donut Sales', size = 'larger', weight = 'bold')
plt.text(x = 0, y = 39.05, s ='The amount sold in each branch for every donut flavor', size = 'large')
plt.xlabel('Sales Branch')
plt.ylabel('Amount of donuts sold')
```
