```
my_list=[1,2,3,4,5]
#write a for loop to square each number in the list
squared=[]
for num in my_list:
    squared.append(num**2)
print(squared)
```

```
squared=[num**2 for num in my_list]
print(squared)
```

```
# do the same for new list with string
#view which error occurs
my_list=[1,2,3,4,5,'6']
squared=[num**2 for num in my_list]
print(squared)
```

```
#create try + except block to handle error
squared=[]
for num in my_list:
    try:
        squared.append(num**2)
    except TypeError:
        squared.append(int(num**2))
print(squared)
```
