```python
#3 Convert to lambda function
def calculate_cube_root(x):
    return x**(1/3)

cube_root = lambda x:x**(1/3)   
```

```python
#4
#call the lambda on the number 64 and round the answer to two decimal places.
round(cube_root(64), 2)
# Answer to 3, 4 in one lambda
round((lambda x:x**(1/3))(64), 2)
```
